```python
import numpy as np
import pandas as pd
import seaborn  as sns
from matplotlib import pyplot as  plt

import plotly.graph_objs as go
import plotly.express as px
```

```python
df= pd.read_csv("jobs_in_data.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
unique_items =df['experience_level'].unique()
print(unique_items)
```

```python
lables =df['experience_level'].value_counts().index
print(lables)
```

```python
values = df['experience_level'].value_counts()
print(values)
```

```python

plt.pie(values,labels=lables,rotatelabels=False,autopct='%1.1f%%',shadow=True,startangle=45)

plt.title("states",color="red",fontsize=15)
```

```python

print(df['work_year'].unique())
print()
print(df['work_year'].value_counts())
```

```python
plt.figure(figsize=(16,10))

sns.countplot(data=df,x='work_year',palette="icefire")
plt.xticks(fontsize=30,rotation=50)

plt.xlabel("job year",fontsize=20,color='black')
plt.ylabel("count",fontsize=20)
plt.title("working year",fontsize=30)
plt.legend()
```

```python
print(df['job_category'].unique())
print()
print(df['job_category'].value_counts())
```

```python
plt.figure(figsize=(20,20))
sns.countplot(data=df,x='job_category',palette="icefire")

plt.xticks(fontsize=15,rotation=50)


plt.xlabel("job_categories")
plt.ylabel("count")
plt.title("numbers of employee in differnt job categories")
```

```python
z=df.groupby('job_category',as_index=False)['salary_in_usd'].max().sort_values(by='salary_in_usd',ascending=False)
print(z)
```
